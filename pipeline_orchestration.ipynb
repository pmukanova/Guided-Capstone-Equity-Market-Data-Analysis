```
%sql

CREATE TABLE IF NOT EXISTS job_tracker_table
  (job_id string , status string, update_time timestamp)
```

```
import datetime 

class Tracker(object):
  """
  job_id, status, updated_time
  """
  def __init__(self, jobname):
    self.jobname = jobname


  def assign_job_id(self):
    job_id = self.jobname+"_"+datetime.date.today().strftime('%Y-%m-%d')
    return job_id


  def update_job_status(self, status):
    job_id = self.assign_job_id()
    print("Job ID Assigned: {}".format(job_id))
    update_time = datetime.datetime.now()
    table_name = "job_tracker_table"
    job_sql = spark.sql(f"INSERT INTO job_tracker_table (job_id, status, update_time) values('{job_id}','{status}','{update_time}')")

  
  def get_job_status(self, job_id):
    record = spark.sql(f"SELECT status FROM job_tracker_table where job_id='{job_id}' ORDER BY update_time desc LIMIT 1")
    return record
```
